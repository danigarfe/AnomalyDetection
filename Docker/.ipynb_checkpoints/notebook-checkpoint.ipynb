<h1>Network Anomaly Detection</h1>

<h3>Importing libraries</h3>

```python
import pandas as pd
import os
from sklearn.svm import LinearSVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.metrics import recall_score, fbeta_score, roc_auc_score, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3>Loading training dataset and initial exploration</h3>

```python
df = pd.read_csv("./Train.txt",sep=",",names=["duration","protocoltype","service","flag","srcbytes","dstbytes","land", "wrongfragment","urgent","hot","numfailedlogins","loggedin", "numcompromised","rootshell","suattempted","numroot","numfilecreations", "numshells","numaccessfiles","numoutboundcmds","ishostlogin",
"isguestlogin","count","srvcount","serrorrate", "srvserrorrate",
"rerrorrate","srvrerrorrate","samesrvrate", "diffsrvrate", "srvdiffhostrate","dsthostcount","dsthostsrvcount","dsthostsamesrvrate", "dsthostdiffsrvrate","dsthostsamesrcportrate",
"dsthostsrvdiffhostrate","dsthostserrorrate","dsthostsrvserrorrate",
"dsthostrerrorrate","dsthostsrvrerrorrate","attack", "lastflag"])
df.head()
```

```python
df.dtypes
```

We can see that the "attack" column indicates the cyberattack type for each dataframe entry. We will process the dataset to convert it to binary (1 if there is an attack, 0 if there is no attack):

```python
df['attack'].loc[df['attack']!='normal']='attack'
```

```python
df[df['attack']!='normal']
```

<h3>Feature encoding</h3>
We will encode the categorical features into numeric ones, with values going from 0 to n_categories-1.

```python
le=LabelEncoder()
df['protocoltype']=le.fit_transform(df['protocoltype'])
df['service']=le.fit_transform(df['service'])
df['flag']=le.fit_transform(df['flag'])
df['attack']=le.fit_transform(df['attack'])
```

<h3>Features cross correlation</h3>
We will check the cross correlation between the features of the dataset. In this way we can detect important features and dataset further corrections. 

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr())
```

```python
df["numoutboundcmds"].describe()
```

We can see that for the "numoutboundcmds" feature all values are 0. We can remove it.

```python
df = df.drop('numoutboundcmds', axis=1)
```

<h3>Features standarization</h3>
We will separate the dataset into input and output dataframes and standarize all input feature values to prevent one variable from having more "weight" than another:

```python
X=df.drop(['attack'],axis=1)
y_train=df['attack']
SS = StandardScaler()
SS.fit(X)
X_train = SS.transform(X)
```

```python
y_train.value_counts()
```

We can see that the dataset is more or less balanced. 

<h3>Loading and processing testing dataset</h3>

```python
df_test = pd.read_csv("./Test.txt",sep=",",names=["duration","protocoltype","service","flag","srcbytes","dstbytes","land", "wrongfragment","urgent","hot","numfailedlogins","loggedin", "numcompromised","rootshell","suattempted","numroot","numfilecreations", "numshells","numaccessfiles","numoutboundcmds","ishostlogin",
"isguestlogin","count","srvcount","serrorrate", "srvserrorrate",
"rerrorrate","srvrerrorrate","samesrvrate", "diffsrvrate", "srvdiffhostrate","dsthostcount","dsthostsrvcount","dsthostsamesrvrate", "dsthostdiffsrvrate","dsthostsamesrcportrate",
"dsthostsrvdiffhostrate","dsthostserrorrate","dsthostsrvserrorrate",
"dsthostrerrorrate","dsthostsrvrerrorrate","attack", "lastflag"])
df_test['attack'].loc[df_test['attack']!='normal']='attack'
df_test['protocoltype']=le.fit_transform(df_test['protocoltype'])
df_test['service']=le.fit_transform(df_test['service'])
df_test['flag']=le.fit_transform(df_test['flag'])
df_test['attack']=le.fit_transform(df_test['attack'])
df_test = df_test.drop('numoutboundcmds', axis=1)
X_2=df_test.drop(['attack'],axis=1)
y_test=df_test['attack']
SS = StandardScaler()
SS.fit(X_2)
X_test = SS.transform(X_2)
```

<h2>SVC Model</h2>

```python
#Training
svm_model = LinearSVC()
svm_model.fit(X_train, y_train)
```

```python
#Testing
y_pred = svm_model.predict(X_test)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print('<<SVC>> Number of false negatives: %d' % (fn))
print('<<SVC>> Misclassified samples: %d' % (y_test != y_pred).sum())
print('<<SVC>> Recall score: %f' % (recall_score(y_test, y_pred)))
print('<<SVC>> F2 Score: %f' % (fbeta_score(y_test, y_pred, beta=2)))
print('<<SVC>> AUC: %f' % (roc_auc_score(y_test, y_pred)))
print('<<SVC>> Accuracy: %.2f%%' % (100.0 * accuracy_score(y_test, y_pred)))
```

<h2>Neural Network</h2>

```python
#Training
neuralnet_model = MLPClassifier(alpha=0.0001, max_iter=10000)
neuralnet_model.fit(X_train, y_train)
```

```python
#Testing
y_pred = neuralnet_model.predict(X_test)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print('<<Neural Net>> Number of false negatives: %d' % (fn))
print('<<Neural Net>> Misclassified samples: %d' % (y_test != y_pred).sum())
print('<<Neural Net>> Recall score: %f' % (recall_score(y_test, y_pred)))
print('<<Neural Net>> F2 Score: %f' % (fbeta_score(y_test, y_pred, beta=2)))
print('<<Neural Net>> AUC: %f' % (roc_auc_score(y_test, y_pred)))
print('<<Neural Net>> Accuracy: %.2f%%' % (100.0 * accuracy_score(y_test, y_pred)))
```

<h2>KNN</h2>

```python
#Train
knn_model = KNeighborsClassifier(metric='minkowski', n_neighbors=5)
knn_model.fit(X_train, y_train)
```

```python
#Test
y_pred = knn_model.predict(X_test)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print('<<KNN>> Number of false negatives: %d' % (fn))
print('<<KNN>> Misclassified samples: %d' % (y_test != y_pred).sum())
print('<<KNN>> Recall score: %f' % (recall_score(y_test, y_pred)))
print('<<KNN>> F2 Score: %f' % (fbeta_score(y_test, y_pred, beta=2)))
print('<<KNN>> AUC: %f' % (roc_auc_score(y_test, y_pred)))
print('<<KNN>> Accuracy: %.2f%%' % (100.0 * accuracy_score(y_test, y_pred)))
```

<h2>Decision Tree</h2>

```python
#Train
tree_model = DecisionTreeClassifier(criterion='entropy', max_depth=10, random_state=1)
tree_model.fit(X_train, y_train)
```

```python
#Test
y_pred = tree_model.predict(X_test)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print('<<Decision Tree>> Number of false negatives: %d' % (fn))
print('<<Decision Tree>> Misclassified samples: %d' % (y_test != y_pred).sum())
print('<<Decision Tree>> Recall score: %f' % (recall_score(y_test, y_pred)))
print('<<Decision Tree>> F2 Score: %f' % (fbeta_score(y_test, y_pred, beta=2)))
print('<<Decision Tree>> AUC: %f' % (roc_auc_score(y_test, y_pred)))
print('<<Decision Tree>> Accuracy: %.2f%%' % (100.0 * accuracy_score(y_test, y_pred)))
```

<h2>Random Forest</h2>

```python
#Train
rf_model = RandomForestClassifier(n_jobs=-1)
rf_model.fit(X_train, y_train)
```

```python
#Test
y_pred = rf_model.predict(X_test)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print('<<Random Forest>> Number of false negatives: %d' % (fn))
print('<<Random Forest>> Misclassified samples: %d' % (y_test != y_pred).sum())
print('<<Random Forest>> Recall score: %f' % (recall_score(y_test, y_pred)))
print('<<Random Forest>> F2 Score: %f' % (fbeta_score(y_test, y_pred, beta=2)))
print('<<Random Forest>> AUC: %f' % (roc_auc_score(y_test, y_pred)))
print('<<Random Forest>> Accuracy: %.2f%%' % (100.0 * accuracy_score(y_test, y_pred)))
```

<h2>Naive Bayes</h2>

```python
#Train
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)
```

```python
#Test
y_pred = gnb_model.predict(X_test)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print('<<NB>> Number of false negatives: %d' % (fn))
print('<<NB>> Misclassified samples: %d' % (y_test != y_pred).sum())
print('<<NB>> Recall score: %f' % (recall_score(y_test, y_pred)))
print('<<NB>> F2 Score: %f' % (fbeta_score(y_test, y_pred, beta=2)))
print('<<NB>> AUC: %f' % (roc_auc_score(y_test, y_pred)))
print('<<NB>> Accuracy: %.2f%%' % (100.0 * accuracy_score(y_test, y_pred)))
```
